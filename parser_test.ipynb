## Comparison of human and bot clicking scatter plots

```python
X_b = bot_game_logs['X-coord']
Y_b = bot_game_logs['Y-coord']
plt.scatter(X_b, Y_b)
plt.title('Bot touch scatterplot')
plt.show()
```

```python
X_h = human_game_logs['X-coord']
Y_h = human_game_logs['Y-coord']
plt.scatter(X_h, Y_h)
plt.title('Human touch scatterplot')
plt.show()
```

### As we can see, bot clicks are more close to each other. Otherwise, human ones are more sparse.
### Also, in bot clicks plot some kind of linerity is observed.

```python
# deviation of clicks for bot
dev_bot = np.std(bot_touch_logs[['X-coord', 'Y-coord']])
print(dev_bot)
# print(np.sqrt(dev_bot.dot(dev_bot)))
print('Norm:', np.linalg.norm(dev_bot))
```

```python
# deviation of clicks for human
dev_human = np.std(human_touch_logs[['X-coord', 'Y-coord']])
print(dev_human)
print('Norm:', np.linalg.norm(dev_human))
```

### Larger deviation norm -> clicks are located more sparsely.
